```python
import pandas as pd

data = {
    "order_id":[101,102,103,104,105,106,107,108,109,110],
    "order_date":["2024-01-01","2024-01-02","2024-01-03","2024-01-04","2024-01-05",
                  "2024-01-06","2024-01-07","2024-01-08","2024-01-09","2024-01-10"],
    "customer_name":["Amit","Ravi","Sneha","Neha","Arjun","Pooja","Rahul","Simran","Karan","Anita"],
    "city":["Delhi","Jaipur","Delhi","Mumbai","Delhi","Jaipur","Mumbai","Delhi","Jaipur","Mumbai"],
    "product":["Laptop","Shirt","Mobile","Shoes","Tablet","Jeans","Laptop","Mobile","Shoes","T-shirt"],
    "category":["Electronics","Fashion","Electronics","Fashion","Electronics","Fashion","Electronics","Electronics","Fashion","Fashion"],
    "quantity":[1,2,1,3,1,2,1,2,1,3],
    "price":[50000,1000,20000,2000,30000,1500,55000,18000,2500,700],
    "status":["Completed","Completed","Cancelled","Completed","Completed","Completed","Completed","Completed","Cancelled","Completed"]
}

df = pd.DataFrame(data)
df.head()
```

```python
df['total_sales']=df['quantity']*df['price']
df.head()
```

```python
# INSIGHTS
```

```python
# 1.TOTAL REVENUE
print("total revenue:",df['total_sales'].sum())
```

```python
# 2.city wise sales
city_sales=(df.groupby('city')['total_sales'].sum())
print(city_sales)
```

```python
# 3.category-wise sales
category_sales=df.groupby('category')['total_sales'].sum()
print(category_sales)
```

```python
# 4.top products sold
top_products=df.groupby('product')['quantity'].sum().sort_values(ascending=False)
print(top_products)
```

```python
# 5.cancelled order and completed order
print(df['status'].value_counts())
```

```python
# 6.highest sales according to city
df.groupby('city')['total_sales'].sum().sort_values(ascending=False)
```

```python
# 7.highest revenue by product
df.groupby('product')['total_sales'].sum().sort_values(ascending=False)
```

```python
# 8.LOSS BY CANCELLED ORDERS
df[df['status'] == 'Cancelled']['total_sales'].sum()
```

```python
# 9.daily sales trend
df.groupby('order_date')['total_sales'].sum()
```

```python
# top 3 customers on revenue basis
df.groupby('customer_name')['total_sales'].sum().sort_values(ascending=False).head(3)
```

```python
df.groupby('city')['total_sales'].sum().sort_values(ascending=False)
```

```python
df[df['status']=='cancelled']['total_sales'].sum()
```

```python
df.groupby('customer_name')['total_sales'].sum().sort_values(ascending=False).head(3)
```

```python
# percentage of cancelled orders
cancel_percent=(df['status'].value_counts(normalize=True)*100)
print(cancel_percent)
```

```python
## Key Insights

# - Delhi generated the highest revenue among all cities.
# - Rahul is the top customer contributing maximum sales.
# - Total loss due to cancelled orders is ₹22,500.
# - 20% of orders were cancelled, indicating a potential area for improvement.
# - Electronics category shows strong performance compared to others.
```
